```python
#Given Valid plans
def validate_recharge(amount):
  valid_plans=[199,299,399,599]
  #To validate the right valid plan
  if amount<50:
    return "Invalid amount. Minimum recharge is ₹50"
  if amount in valid_plans:
    return "Recharge successful"
  else:
    return "Invalid plan amount"
#Using While loop to continuously allow user retry until successful recharge
while True:
  amount=int(input("Enter recharge amount:"))
  result=validate_recharge(amount)
  print(result)
  if result=="Recharge successful":
    break
  retry=input("Retry? (yes/no):")
  if retry=="no":
    break
```

```python
def check_inventory(products):
  #Looping through each product in warehouse inventory database
  for product in products:
    stock=products[product]
    #To detect low stock condition and alert for reordering
    if stock<15:
      print(product,"→ Reorder Alert")
    else:
      print(product,"→ Stock OK")
#Given inventory list
products={
"Laptop":10,
"Mouse":25,
"Keyboard":12,
"Monitor":30
}
check_inventory(products)
```

```python
def process_result(marks):
  total=0
  #Looping through each marks to out average
  for mark in marks:
    total=total+mark
    average=total/len(marks)
  if average>=50:
    return "Pass"
  else:
    return "Fail"
#Given marks
marks=[65,70,45,80,55]
result=process_result(marks)
print("Result:",result)
```

```python
#Function to calculate cab fare based on distance and peak hour pricing
def calculate_fare(distance,peak):
  #Base fare calculation
  fare=50+(distance*12)
  #Peak hour calculation
  if peak=="yes":
    fare=fare+(fare*0.25)
  return fare
while True:
  distance=float(input("Enter distance in km:"))
  peak=input("Is it peak hour? (yes/no):")
  fare=calculate_fare(distance,peak)
  print("Total Fare: ₹",fare)
  #User retry for calculating again
  retry=input("Calculate again? (yes/no):")
  if retry=="no":
    break
```

```python
def check_attendance(attendance):
  present=0
  #Loopin through each days for counting present days
  for day in attendance:
    if day=="P":
      present=present+1
  #Calculating percentage of present days
  percentage=(present/len(attendance))*100
  if percentage>=75:
    return "Eligible"
  else:
    return "Not Eligible"
#Given Attendance
attendance=["P","P","A","P","P","A","P","P"]
result=check_attendance(attendance)
print("Attendance Status:",result)
```

```python
def check_password(password):
  #Flags to determine security factors
  has_digit=False
  has_special=False
  special_chars="@#$*/"
  #Check the length of passwords and if falls below it reject it
  if len(password)<8:
    return "Weak Password"
  #To check existence of special characters and digits
  for char in password:
    if char.isdigit():
      has_digit=True
    if char in special_chars:
      has_special=True
  if has_digit and has_special:
    return "Strong Password"
  else:
    return "Weak Password"
password=input("Enter password:")
result=check_password(password)
print(result)
```

```python
#Given Sentence
Sents=input("Enter the Sentence:")
#Splitting the sentence using Split function
Single_words=Sents.split()
#Using Set function to remove duplicates and finding unique words
Set_words=set(Single_words)
#Printing the Count and Set of unique words
print(f"Count of words: {len(Set_words)}, Set of Unique Words: {Set_words}")
```

```python
#Given Sentence
Sentence=input("Enter sentence:")
#Splitting the sentence using Split function
Words=Sentence.split()
#To print in order we manually loop through each words to find unique words
Unique_words=[]
for word in Words:
    if word not in Unique_words:
        Unique_words.append(word)
print(f"Unique Words Count: {len(Unique_words)}\nUnique Words: {Unique_words}")
```

```python
#Given Employees with salaries
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000}
#Intializing the highest salary to 0 and name to none
highest_salary=0
highest_employee=None
#Loop through each employee to find the highest salary by comparison
for name in employees:
  #The first employee will have the highest salary at the start
  salary=employees[name]
  #Comparing each employee salary to its neighboring employee salary and switching the highest salary variable
  if salary > highest_salary:
    highest_salary=salary
    highest_employee=name
print(f"Highest Salary: {highest_employee} - {highest_salary}")
```

```python
#Given Attendance
attendance = ["P", "P", "A", "P", "P"]
#Counter Variable for present days
present_days=0
#Total no of days using len function
total_days=len(attendance)
#Loop through each days to find present and increment it
for days in attendance:
  if days=="P":
    present_days=present_days+1
print("Attendance Percentage:",present_days/total_days*100)
```

```python
numbers = [45, 22, 89, 10, 66]
#Initializing both maximum and minimum with the first element of the list.
maximum=numbers[0]
minimum=numbers[0]
#Looping through each element for comparison
for num in numbers:
  if num>maximum:
    maximum=num
  if num<minimum:
    minimum=num
#Printing the results
print("List:",numbers)
print("Max:", maximum)
print("Min:",minimum)
```

```python
#Given Text
text="encyclopedia"
#Empty Dictionary
freq={}
#Loop through each letter in text to find the frequency and increment it
for letter in text:
  if letter in freq:
    freq[letter]+=1
  else:
    freq[letter]=1
print(freq)
```

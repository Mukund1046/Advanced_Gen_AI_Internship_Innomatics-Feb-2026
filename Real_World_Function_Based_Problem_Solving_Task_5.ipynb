```python
#Function to calculate total likes and determine engagement level
def analyze_engagement(likes):
  total=0
  #Loop through each like count to calculate total likes
  for like in likes:
    total=total+like
  #Checking if post is viral or normal
  if total>=1000:
    status="Viral Post"
  else:
    status="Normal Engagement"
  print("Total Likes:",total)
  print("Post Status:",status)
#Given likes data from different sources
likes=[200,300,150,400]
analyze_engagement(likes)
```

```python
#Function to check medicine stock level
def check_stock(stock):
  #Checking if stock is below minimum level
  if stock<10:
    status="Low Stock Alert"
  else:
    status="Stock Sufficient"
  print("Medicine Stock:",stock)
  print("Status:",status)
#Given stock value
stock=6
check_stock(stock)
```

```python
#Function to calculate average rainfall and check adequacy
def check_rainfall(rainfall,required):
  total=0
  #Loop through rainfall data to calculate total rainfall
  for rain in rainfall:
    total=total+rain
  average=total/len(rainfall)
  #Checking rainfall adequacy
  if average>=required:
    status="Adequate Rainfall"
  else:
    status="Inadequate Rainfall"
  print("Average Rainfall:",average)
  print("Rainfall Status:",status)
#Given rainfall data
rainfall=[60,75,80,70,75]
required=70
check_rainfall(rainfall,required)
```

```python
#Function to detect duplicate usernames
def detect_duplicates(usernames):
  unique=set()
  duplicate_found=False
  #Loop through usernames to check duplicates
  for user in usernames:
    if user in unique:
      duplicate_found=True
    else:
      unique.add(user)
  if duplicate_found:
    print("Duplicate Accounts Found: Yes")
  else:
    print("Duplicate Accounts Found: No")
#Given username list
usernames=["mukund","rahul","ananya","mukund"]
detect_duplicates(usernames)
```

```python
#Function to check appointment eligibility based on age
def check_eligibility(age):
  #Checking eligibility condition
  if age>=18:
    status="Eligible"
  else:
    status="Not Eligible"
  print("Patient Age:",age)
  print("Eligibility Status:",status)
#Given age
age=21
check_eligibility(age)
```

```python
#Function to filter premium crop prices
def filter_premium(prices):
  premium=[]
  #Loop through prices and filter premium crops
  for price in prices:
    if price>2000:
      premium.append(price)
  print("Premium Crops:",premium)
#Given crop prices
prices=[1500,2500,1800,3200,900]
filter_premium(prices)
```

```python
#Function to check system health based on error count
def check_health(errors):
  #Checking system condition based on errors
  if errors==0:
    status="Healthy"
  elif errors<=5:
    status="Minor Issues"
  else:
    status="Critical Issues"
  print("Error Count:",errors)
  print("System Status:",status)
#Given error count
errors=7
check_health(errors)
```

```python
#Function to validate transaction amount
def check_transaction(amount):
  limit=50000
  #Checking if transaction is within limit
  if amount<=limit:
    status="Approved"
  else:
    status="Rejected"
  print("Transaction Amount:",amount)
  print("Transaction Status:",status)
#Given transaction amount
amount=60000
check_transaction(amount)
```

```python
#Function to calculate attendance percentage and eligibility
def check_attendance(attendance):
  present=0
  #Loop through attendance to count present days
  for day in attendance:
    if day=="P":
      present=present+1
  percentage=(present/len(attendance))*100
  #Checking exam eligibility
  if percentage>=75:
    status="Eligible"
  else:
    status="Not Eligible"
  print("Attendance Percentage:",percentage)
  print("Exam Eligibility:",status)
#Given attendance record
attendance=["P","P","A","P","P","P","A","P","P","P"]
check_attendance(attendance)
```

```python
#Function to calculate electricity bill and usage status
def calculate_bill(units):
  bill=0
  #Calculating bill based on slab rates
  if units<=100:
    bill=units*3
  elif units<=200:
    bill=(100*3)+((units-100)*5)
  else:
    bill=(100*3)+(100*5)+((units-200)*7)
  #Checking usage classification
  if bill<500:
    status="Low Usage"
  elif bill<=1500:
    status="Moderate Usage"
  else:
    status="High Usage"
  print("Units Consumed:",units)
  print("Electricity Bill:",bill)
  print("Usage Status:",status)
#Given units consumed
units=250
calculate_bill(units)
```

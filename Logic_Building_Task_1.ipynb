```python
#Given Username and Password
Username="Mukund"
Password="123456"
#Input Credentials
name=input("Enter Your name:")
password=input("Enter Your Password:")
#Condition to check if both username and password match
if Username==name and Password==password:
  print("Login Successful")
else:
  print("Invalid Credentials")
```

```python
#Given marks of students
marks = [45, 78, 90, 33, 60]
#Counter Variables
Passed_students=0
Failed_Students=0
#Looping Through each Student Marks
for mark in marks:
  if mark>=50:
    Passed_students+=1
  else:
    Failed_Students+=1
#Using len function to count the total number of students
print("Total Students:",len(marks))
print("Passed Students:",Passed_students)
print("Failed Students:",Failed_Students)
```

```python
#Given Names
names = [" Alice ", "bob", " CHARLIE "]
#Empty list to store processed names
Standardized_names=[]
#Loop Through each names(Using strip to remove trailing spaces and lower function to Lowercase the names)
for name in names:
    name=name.strip()
    name=name.lower()
    Standardized_names.append(name)
print(Standardized_names)
```

```python
#Given Messages
messages = ["Hi", "Welcome to the platform", "OK"]
#Loop Through each messages and find its length using len function
for message in messages:
    length=len(message)
    print(f"Message: {message}, Length:{ length}")
    #Condition to check if the message length is greater than 10 characters and flag it
    if len(message)>10:
      print("The Above Message is longer than 10 characters")
```

```python
#Given System logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
#Counter Variable
err_count=0
#Loop through each logs and increment the error count
for log in logs:
  if log=="ERROR":
    err_count+=1
#Print the Total Error Count
print("Total Error Count:",err_count)
```

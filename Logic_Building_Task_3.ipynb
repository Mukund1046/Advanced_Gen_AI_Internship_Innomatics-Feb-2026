```python
#Given Data
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
    }
#Manually find highest score by comaprison
highest_score=0
for name in employees:
    score = employees[name]
    if highest_score<score:
        highest_score =score
    elif score > highest_score:
        highest_score=score
#Collect employees who match highest score
top_performers=[]
for name in employees:
    if employees[name]==highest_score:
        top_performers.append(name)
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), "(Score:", highest_score, ")")
```

```python
#Given query
query="Buy mobile phone buy phone online"
# Convert to lowercase manually
query=query.lower()
#Removing punctuation using alnum
cleaned_query=""
for char in query:
    if char.isalnum() or char==" ":
        cleaned_query += char
#Split words
words=cleaned_query.split()
#Frequency counting
freq={}
for word in words:
    if word in freq:
        freq[word]=freq[word] + 1
    else:
        freq[word]=1
#Print only words appearing more than once
result={}
for word in freq:
    if freq[word]>1:
        result[word]=freq[word]
print(result)
```

```python
#Given Sensor readings
sensor_readings=[3, 4, 7, 8, 10, 12, 5]
valid_readings=[]
index=0
#Loop through the readings to find valid readings which are even numbers
for value in sensor_readings:
    if value%2==0:
        valid_readings.append((index, value))
    index=index + 1
print("Valid Sensor Readings (Hour, Value):\n",valid_readings)
```

```python
#Given emails
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
    ]
domain_count={}
#Loop through each emails and find the domain names using slice operator and increment the count
for email in emails:
    at_index=email.index("@")
    domain=email[at_index+1:]
    if domain in domain_count:
        domain_count[domain]=domain_count[domain]+1
    else:
        domain_count[domain]=1
#Total emails
total=len(emails)
for domain in domain_count:
    percentage=(domain_count[domain]/total)*100
    print(domain + ":",int(percentage),"%")
```

```python
#Given sales data
sales = [1200, 1500, 900, 2400, 1400, 3000]
# Calculating sum
total=0
for value in sales:
    total+=value
average=total/len(sales)
#Detecting sales spike where sales are more than 30% above average
threshold=average+(0.3*average)
day=1
for value in sales:
    if value>threshold:
        print("Day", day, ":", value)
    day=day+1
```

```python
#Given user id's
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
id_count={}
#Count the frequency of recurring user id's
for user in user_ids:
    if user in id_count:
        id_count[user]=id_count[user]+1
    else:
        id_count[user]=1
#To print number of duplicate id's if it has appeared more than once
for user in id_count:
    if id_count[user]>1:
        print(user, "→", id_count[user], "times")
```
